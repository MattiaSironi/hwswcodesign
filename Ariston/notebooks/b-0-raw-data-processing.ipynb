```python
from pathlib import Path

import pandas as pd
import seaborn as sns
```

```python
DATA_DIR = Path("../data/")
RAW_DATA_DIR = DATA_DIR / "0_ariston" / "Raw_Data_Collection"
PROCESSED_DATA_DIR = DATA_DIR / "1_processed"

DATASET_DIR = RAW_DATA_DIR / "Boiler2HE - dataset" / "Hungary" / "AD2740AC6A29_ar1_Log_log"
DATASET_DIR
```

```python
raw_dfs = [
    (
        pd.read_csv(path, sep=";")
        .dropna(axis="columns", how="all")
        .dropna(axis="index", how="all")
        .drop(["Topic", "BodyBytes"], axis="columns")
        .convert_dtypes()
    )
    for path in DATASET_DIR.glob("*.csv")
]
raw_df = pd.concat(raw_dfs)
raw_df["Timestamp"] = pd.to_datetime(raw_df["Timestamp"])
raw_df["datetime"] = raw_df["Timestamp"]
raw_df = raw_df.set_index("Timestamp").sort_index()
raw_df.info()
```

```python
raw_df.index
```

Reading column names

```python
legend_df = pd.read_excel(RAW_DATA_DIR / "DGTO_Legend.xlsx", header=None)
legend_df
```

```python
boiler_legend_df = (
    legend_df.iloc[13:]
    .T
    .dropna(how="all")
    .drop(0)
)
boiler_legend_df.columns = ["id", "name", "description"]
boiler_legend_df
```

```python
with open(PROCESSED_DATA_DIR / "boiler_dict.txt", mode="w", encoding="utf8") as f:
    for _, _, name, description in boiler_legend_df.itertuples():
        name_description = str(name) + "\n\t" + str(description)
        f.write(name_description + "\n")
        print(name_description)
```

```python
id_name_map = {
    row["id"]: row["name"].strip()
    for _, row in boiler_legend_df.iterrows()
}
id_name_map
```

```python
df = raw_df.rename(columns=id_name_map)
df.info()
```

```python
df.to_parquet(PROCESSED_DATA_DIR / "hungary.parquet")
```
